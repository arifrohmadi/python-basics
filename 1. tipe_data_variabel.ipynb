```python
# Variables

# Assign value ke variabel

#1 basic assignment
x = 7
y = 3.14
z = "Hai"
print(x,y,z)

#2 dynamic typing
a = 10
a = "Solo"
print(a)
```

```python
# multiple assignment

#1 assign nilai yang sama
x = y = z = 80
print(x,y,z)

#2 assign nilai yang berbeda
a, b, c = 1, 3.14, "Python"
print(a, b, c)
```

```python
# Casting variables
s = "8"  # Initially a string
n = int(s)  # Cast string to integer
cnt = 7
f = float(cnt)  # Cast integer to float
age = 20
s2 = str(age)  # Cast integer to string

# Display results
print(n)
print(f)
print(s2)
```

```python
# Scope of Variable

# Variabel lokal
def f():
    a = "Saya variabel lokal"
    print(a)

f()
print(a)
```

```python
# Variabel global

a = "Saya variabel global"

def f():
    global a
    a = "Variabel global dimodifikasi"
    print(a)

f()
print(a)
```
