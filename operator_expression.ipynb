```python
# operator dan ekspresi

#1.operator aritmatika
a = 11
b = 5

# Addition
print("Addition:", a + b)

# Subtraction
print("Subtraction:", a - b)

# Multiplication
print("Multiplication:", a * b)

# Division
print("Division:", a / b)

# Floor Division
print("Floor Division:", a // b)

# Modulus
print("Modulus:", a % b)

# Exponentiation
print("Exponentiation:", a ** b)
```

```python
# Concatenation and Repetition
print("Halo, " + "Dunia!")

combine = ("a", "b", "c", "d") + ("E", "F")
print(combine)

combine2 = [0, 1, 2, 3] + [4, 5, 6,7]
print(combine2)

repeat = "Halo " * 3
print(repeat)

repeat2 = 3 * [1,2,3]
print(repeat2)
```

```python
#2.operator penugasan
a = 5
b = a
print(b)

b += a # equal with b = b+a
print(b)
b -= a # equal with b = b-a
print(b)
b *= a # equal with b = b*a
print(b)
b /= a # equal with b = b/a
print(b)
```

```python
#3.operator perbandingan
a = 11
b = 40

print(a > b)
print(a < b)
print(a == b)
print(a != b)
print(a >= b)
print(a <= b)
```

```python
#4.operator logika
a = True
b = False
print(a and b)
print(a or b)
print(not a)
```

```python
#5.operator identitas
x = 1001
y = 1001

print(x == y)
print(x is y)
print(id(x))
print(id(y))
print("")

a = "halo"
b = a
print(a is b)
print(id(a))
print(id(b))
```

```python
#6.operator keanggotaan
angka = [1,3,5,7,9]
a = 5
print(a in angka)
b = 2
print (b not in angka)

text = "Merdeka"
c = "m"
print (c in text)
```

```python
#7.operator bitwise
print(10 & 7)
print(10 | 7)
print(10 ^ 7)
print(~10)
print("")
print(10 << 1)
print(7 >> 1)
```
